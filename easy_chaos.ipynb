```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')

def get_seq(x, n=100):
    ans = []
    for i in range(n):
        x = x**2 - 2
        ans.append(x)
    return ans
x = 0.4
a = get_seq(0.4)
b = get_seq(0.4001)
plt.plot(a, color='red', linestyle='--');
plt.plot(b, color='blue');
```
