```python
import requests
from lxml import html
from tqdm import tqdm_notebook as tqdm
```

```python
def get_target_row(tree, target_tag):
    target_element_list = tree.xpath(f".//*[contains(text(),'{target_tag}')]")    
    ans_list = []
    
    
    for e in target_element_list:
        cur_type = None
        if str(e).startswith("<Element td "):
            cur_type = "td"
            target_tr = e.getparent()
        elif str(e).startswith("<Element span "):
            # span case
            cur_type = "span"
            if str(e.getparent()).startswith("<Element div "):
                target_tr = e.getparent()
                cur_type = "div"
            else:
                target_tr = e.getparent().getparent()
                if not str(target_tr).startswith("<Element tr "):
                    cur_type = "undefined"
            
            
        if cur_type in ["span", "td", "div"]:
            cur_ans_list = [td.text if td.text else td.find("span").text for td in target_tr]
            ans_list.append(cur_ans_list)
        
    return ans_list

def get_site_target(target, target_tag):
    headers: dict = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = requests.get(target, headers=headers)
    tree = html.fromstring(response.content)
    for row in get_target_row(tree, target_tag):
        print('\t', row)
```
