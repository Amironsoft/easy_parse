```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
timerange1 = pd.date_range('2018-01-01 00:00:00', '2018-01-01 22:00:00', freq='2h')
var1 = ['ночь', 'ночь', 'ночь', 'ночь', 'день', 'день', 'день',  'день',  'вечер', 'вечер', 'вечер', 'вечер']
var2 = [82.87, 99.25, 89.82, 80.51, 80.11, 86.91, 90.71, 80.06, 84.51, 84.7, 83.9, 80.21]
var3 = [86.88, 103.28, 96.28, 86.8, 84.86, 92.21, 94.25, 86.1, 89.83, 92.5, 92.43, 86.6]
df = pd.DataFrame({"Smena": var1, "AL2O3_M1": var2, "NA2O_M1":var3}, index=timerange1)
```

```python
df.head()
```

```python
df.plot();
```

```python
df_interp = df.resample('1min').interpolate(method='cubic', order=2)
df_interp.head()
```

```python
df_interp.plot();
```

```python
df_interp.to_csv("interpolate_example_cubic.csv", sep=';', decimal=',')
```
